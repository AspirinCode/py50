# Plot in Grid

The dose-response curves can quickly be plotted in a grid format. This is achieved using the grid_curve_plot() method.

First, import the necessary packages. Import the example dataset and instantiate it into the PlotCurve() class.

```python
import pandas as pd
from py50.plotcurve import PlotCurve
from py50.plot_settings import CBMARKERS, CBPALETTE
```

```python
df = pd.read_csv('../dataset/multiple_example.csv')
data = PlotCurve(df)
data.show().head()
```

The grid_curve_plot() method has many arguments. Here we will utilize the basic methods to generate our first figure. By default, it will generate a 4 X 4 grid. 

```python
figure = data.grid_curve_plot(name_col='Compound Name',
                              concentration_col='Compound Conc',
                              response_col='% Inhibition Avg')
```

We can reposition how the plots are laid out using the "column_num=" argument. Note that if this argument is called, the figures will be "distroted". The subplots can be adjusted by including the "figsize=" argument and adjusting the size accordingly.

```python
figure = data.grid_curve_plot(name_col='Compound Name',
                              concentration_col='Compound Conc',
                              response_col='% Inhibition Avg',
                              column_num=4,
                              figsize=(15,4))
```

The figures can be further modified to add a title, adjust the line colors, and highlight the Absolute IC50 for each drug. 

```python
figure = data.grid_curve_plot(name_col='Compound Name',
                              concentration_col='Compound Conc',
                              response_col='% Inhibition Avg',
                              plot_title='Four Drug Example',
                              line_color=['#FA8072', '#BDB76B', '#D8BFD8', '#808000'],
                              box=True,
                              box_color='red')
```
